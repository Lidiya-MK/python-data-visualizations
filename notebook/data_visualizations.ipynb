# Filtering the Dataset

- Checking the different values there are for a specific column

```python
car_financing_dataframe[["car_type"]].value_counts()
```

- Filtering data for a particular car_type

```python
#Let's say we just want VW Golf R type cars from the dataset

vw_filter= car_financing_dataframe['car_type']=="VW Golf R"

```

```python
vw_filter.head()
```

```python
car_financing_dataframe.loc[vw_filter,:][0:10]
```

- Changing the old dataframe to the new dataframe with car_type of VW Golf R

```python
car_financing_dataframe= car_financing_dataframe.loc[vw_filter,:]
```

```python
car_financing_dataframe
```

- Further filtering the data by setting an Interest Paid boundary

```python

ip_filter= car_financing_dataframe['Interest Paid'] >=100.0
```

```python
ip_filter.head()
```

```python
car_financing_dataframe.loc[ip_filter,:]
```

```python
car_financing_dataframe["Interest Paid"].value_counts(dropna=False)
```

- Applying multiple filters to a dataset simultaneously

```python
#lets create filters for repayment above 1000 and month above 10

repayment_filter= car_financing_dataframe['Repayment']>=1000
months_filter= car_financing_dataframe['Month']>=10
```

```python
car_financing_dataframe= car_financing_dataframe.loc[repayment_filter & months_filter,:]
```

```python
car_financing_dataframe
```

**Removing unnecessary Data**

```python
# first I will rename a few columns I noticed have spacing in their names
car_financing_dataframe= car_financing_dataframe.rename(columns={
    "Interest Paid":"interest_paid",
    "Principal Paid":"principal_paid",
    "New Balance":"new_balance",
    "Starting Balance":"starting_balance"
})

# It is possible to also use car_financing_dataframe.columns=[list of all columns with their new names or old names if they are not meant to change]
# But I prefer this since it is less error prone 
```

```python
car_financing_dataframe
```

- lets remove some columns


```python
car_financing_dataframe= car_financing_dataframe.drop(columns='term')

# del car_financing_dataframe['term'] --- also possible
```

```python
car_financing_dataframe
```

**Cumulative Interest Paid for High-Payment, Long-Term Loans on a Specific Model**
